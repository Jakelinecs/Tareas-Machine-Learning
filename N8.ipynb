```python
import numpy as np
import matplotlib.pyplot as plt

x_linear = np.arange(-50, 50.1, 0.1)

y_linear = (1/2) * x_linear + 1

array_xy_linear = np.stack((x_linear, y_linear), axis=1)

print("--- 1次関数 (y = 0.5x + 1) ---")
print(f"xの要素数: {len(x_linear)}")
print(f"array_xyの形状: {array_xy_linear.shape}")
print(f"先頭5行のデータ:\n{array_xy_linear[:5]}")
```

```python
##############################################################################################################
```

```python
delta_y_linear = np.diff(y_linear)

delta_x_linear = np.diff(x_linear)

gradient_linear = delta_y_linear / delta_x_linear

print("\n勾配の計算結果 (dy/dx):")
print(f"勾配の要素数: {len(gradient_linear)}")
print(f"勾配の例 (先頭5つ): {gradient_linear[:5]}")
```

```python
#############################################################################################################
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x_linear, y_linear, color='blue', label='y = 0.5x + 1')
plt.title('1次関数のグラフ')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle='--')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x_linear[:-1], gradient_linear, color='red', label='Gradiente')
plt.title('1次関数の勾配')
plt.xlabel('x')
plt.ylabel('dy/dx')
plt.axhline(0.5, color='gray', linestyle='-')
plt.ylim(0, 1)
plt.grid(True, linestyle='--')
plt.legend()

plt.tight_layout()
plt.show()
```

```python

def function1(array_x):
    """y = x²"""
    return array_x ** 2

def function2(array_x):
    """y = 2x² + 2x"""
    return 2 * (array_x ** 2) + 2 * array_x

def function3(array_x):
    """y = sin(x/2)"""
    return np.sin(array_x / 2)


def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    x = np.arange(*x_range)
    y = function(x)

    array_xy = np.stack((x, y), axis=1)

    gradient = np.diff(y) / np.diff(x)

    return array_xy, gradient

functions_to_test = [
    (function1, (-50, 50.1, 0.1), "y = x²"),
    (function2, (-50, 50.1, 0.1), "y = 2x² + 2x"),
    (function3, (0, 50.1, 0.1), "y = sin(x/2)")
]

plt.figure(figsize=(15, 12))

print("\n--- 2次関数・三角関数の勾配計算 ---")

for i, (func, x_range, title) in enumerate(functions_to_test):
    array_xy, gradient = compute_gradient(func, x_range)
    x = array_xy[:, 0]
    y = array_xy[:, 1]

    print(f"関数: {title}, x: {x_range[0]}〜{x_range[1]-0.1}, 勾配の形状: {gradient.shape}")

    plt.subplot(3, 2, 2*i + 1)
    plt.plot(x, y, color='blue')
    plt.title(f': {title}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True, linestyle='--')

    # グラフ2 (右): 勾配
    plt.subplot(3, 2, 2*i + 2)
    plt.plot(x[:-1], gradient, color='red')
    plt.title(f' (dy/dx) of {title}')
    plt.xlabel('x')
    plt.ylabel('勾配')
    plt.axhline(0, color='black', linestyle='-', linewidth=0.5) # 勾配ゼロのライン
    plt.grid(True, linestyle='--')

plt.tight_layout()
plt.show()
```

```python
print("\n--- 最小値の探索と勾配の確認 ---")

for i, (func, x_range, title) in enumerate(functions_to_test):
    array_xy, gradient = compute_gradient(func, x_range)
    x = array_xy[:, 0]
    y = array_xy[:, 1]

    min_y = np.min(y)
    index_min = np.argmin(y)
    x_min = x[index_min]

    if index_min > 0:
        gradient_before = gradient[index_min - 1]
    else:
        gradient_before = "N/A (範囲の始まり)"

    if index_min < len(gradient):
        gradient_after = gradient[index_min]
    else:
        gradient_after = "N/A (範囲の終わり)"

    print(f"\n[関数: {title}]")
    print(f"  最小値 y: {min_y:.4f} (x={x_min:.1f})")
    print(f"  最小値のx座標のインデックス: {index_min}")
    print(f"  最小値直前 ({x[index_min-1]:.1f}〜{x_min:.1f}) の勾配: {gradient_before}")
    print(f"  最小値直後 ({x_min:.1f}〜{x[index_min+1]:.1f}) の勾配: {gradient_after}")
```
