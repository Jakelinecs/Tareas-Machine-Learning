```python
import math
import matplotlib.pyplot as plt

RADIUS_MANJU_M = 0.03  # 3 cm en metros
V_MANJU_0 = (4/3) * math.pi * (RADIUS_MANJU_M ** 3)

AU_TO_METERS = 1.5 * (10**11) #     1.5e11
RADIUS_SS_M = 30 * AU_TO_METERS  # 4.5 x 10^12 metros
V_SYSTEM_SOLAR = (4/3) * math.pi * (RADIUS_SS_M ** 3)

TIME_TO_DOUBLE_MIN = 5

print(f"Volumen inicial del manju: {V_MANJU_0:.2e} m췁")
print(f"Volumen del Sistema Solar: {V_SYSTEM_SOLAR:.2e} m췁")
```

```python
def calculate_filling_time(initial_volume, target_volume, time_to_double_min=5):
    current_volume = initial_volume
    total_time_min = 0
    volume_history = [initial_volume]

    # El volumen se duplica 2^n veces, donde n es el n칰mero de duplicaciones.
    while current_volume < target_volume:
        current_volume *= 2
        total_time_min += time_to_double_min
        volume_history.append(current_volume)

    return total_time_min, volume_history

# Calcular el tiempo para llenar el Sistema Solar
total_time_min, volume_history = calculate_filling_time(
    V_MANJU_0, V_SYSTEM_SOLAR, TIME_TO_DOUBLE_MIN
)

# Conversi칩n a horas y d칤as para mejor comprensi칩n
total_time_hr = total_time_min / 60
total_time_day = total_time_hr / 24

print("\n--- Resultados del Manju de Casta침as ---")
print(f"N칰mero de duplicaciones requeridas: {len(volume_history) - 1}")
print(f"Tiempo total para llenar el Sistema Solar: {total_time_min:,} minutos")
print(f"Equivalente en horas: {total_time_hr:,.2f} horas")
print(f"Equivalente en d칤as: {total_time_day:,.2f} d칤as 游뱚")
```

```python
time_steps_min = [i * TIME_TO_DOUBLE_MIN for i in range(len(volume_history))]

plt.figure(figsize=(10, 6))

plt.plot(time_steps_min, volume_history, 'r-o', markersize=3, alpha=0.7)

plt.axhline(V_SYSTEM_SOLAR, color='blue', linestyle='--', label='Volumen del Sistema Solar')

plt.yscale('log')

plt.title("Crecimiento Exponencial del Volumen del Manju de Casta침as", fontsize=16)
plt.xlabel("Tiempo Transcurrido (minutos)", fontsize=12)
plt.ylabel("Volumen (m췁) [Escala Logar칤tmica]", fontsize=12)
plt.legend()
plt.grid(True, linestyle=':')
plt.show()
```

```python
# Objetivo de prueba: Tokyo Dome
V_TOKYO_DOME = 1.24e6  # 1.24 millones de metros c칰bicos

# Usar la funci칩n de prop칩sito general
time_dome_min, _ = calculate_filling_time(
    V_MANJU_0, V_TOKYO_DOME, TIME_TO_DOUBLE_MIN
)

time_dome_hr = time_dome_min / 60

print("\n--- Ejemplo General: Llenar el Tokyo Dome ---")
print(f"Volumen del Tokyo Dome: {V_TOKYO_DOME:,.2f} m췁")
print(f"Tiempo total para llenarlo: {time_dome_min:,} minutos")
print(f"Equivalente en horas: {time_dome_hr:,.2f} horas")
```
