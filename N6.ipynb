```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
```

```python
mean_0 = np.array([-3, 0])
cov_0 = np.array([
    [1.0, 0.8],
    [0.8, 1.0]
])
n_samples = 500

data_0 = np.random.multivariate_normal(mean_0, cov_0, n_samples)

print("--- Pregunta 1: Datos de la Clase 0 ---")
print(f"Forma de los datos de la Clase 0: {data_0.shape}")
print(f"Primeros 5 puntos de datos:\n{data_0[:5]}")
```

```python
##############################################################################################################
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(data_0[:, 0], data_0[:, 1], s=20, alpha=0.6, color='blue', label='Clase 0')
plt.title('Scatter')
plt.xlabel('Dimensión X1')
plt.ylabel('Dimensión X2')
plt.grid(True, linestyle='--')
plt.legend()
plt.show()
```

```python
##############################################################################################################
```

```python
x_min = np.floor(data_0.min())
x_max = np.ceil(data_0.max())

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data_0[:, 0], bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Histograma de la Dimensión 1 (x)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.xlim(x_min, x_max)
plt.grid(axis='y', alpha=0.5)

plt.subplot(1, 2, 2)
plt.hist(data_0[:, 1], bins=30, alpha=0.7, color='lightcoral', edgecolor='black')
plt.title('Histograma de la Dimensión 2 (y)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.xlim(x_min, x_max)
plt.grid(axis='y', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
##############################################################################################################
```

```python
mean_1 = np.array([0, -3])
n_samples = 500
cov_1 = np.array([
    [1.0, 0.8],
    [0.8, 1.0]
])

data_1 = np.random.multivariate_normal(mean_1, cov_1, n_samples)

plt.figure(figsize=(8, 8))

plt.scatter(data_0[:, 0], data_0[:, 1], s=20, alpha=0.6, color='blue', label='0')

plt.scatter(data_1[:, 0], data_1[:, 1], s=20, alpha=0.6, color='#FF8C00', label='1')

plt.title('Diagrama de Dispersión Combinado (Datos Ficticios)', fontsize=14)
plt.xlabel('Dimensión 1 (x)')
plt.ylabel('Dimensión 2 (y)')
plt.grid(True, linestyle='--')
plt.legend(title='Clase')
plt.show()
```

```python
#####################################################################################################################
```

```python
combined_data = np.vstack((data_0, data_1))

print("--- Problema 5: Combinación de Datos ---")
print(f"Forma del ndarray combinado: {combined_data.shape}")
print(f"Primeros 5 puntos:\n{combined_data[:5]}")
print(f"Últimos 5 puntos:\n{combined_data[-5:]}")
```

```python
#####################################################################################################################
```

```python
labels_0 = np.zeros((n_samples, 1))
labels_1 = np.ones((n_samples, 1))

combined_labels = np.vstack((labels_0, labels_1))

final_dataset = np.hstack((combined_data, combined_labels))

print("\n--- 【問題6】ラベル付け ---")
print(f"Forma del conjunto de datos final: {final_dataset.shape}")
print("\nPrimeros 5 puntos (Clase 0):")
print(final_dataset[:5])
print("\nPuntos cerca del límite (Clase 0 -> Clase 1):")
print(final_dataset[498:503])
print("\nÚltimos 5 puntos (Clase 1):")
print(final_dataset[-5:])
```
