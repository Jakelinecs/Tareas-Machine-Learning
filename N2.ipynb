```python
def compute_sorori_shinzaemon(n_days=100):
    """曽呂利新左衛門の米の逸話における、日にちごとの米の粒の数を計算する関数。

    Parameters
    -----------
    n_days: int
        米を貰う日数 (default : 100)

    Returns
    -------
    list_n_grains : list
        ある日に貰う米の数のリスト (リストの長さはn_daysに等しい)
    list_total_grains : list
        ある日までに累計で貰う米の数のリスト (リストの長さはn_daysに等しい)
    """
    list_n_grains = []
    list_total_grains = []
    total_grains = 0

    for day in range(1, n_days + 1):
        # La cantidad de granos en el día 'day' es 2^(day - 1)
        grains_today = 2 ** (day - 1)
        list_n_grains.append(grains_today)

        # Calcular el total acumulado
        total_grains += grains_today
        list_total_grains.append(total_grains)

    # ここにコードを書く。passは何もしないことを意味する命令である。
    pass

    return list_n_grains, list_total_grains

list_n_grains, list_total_grains = compute_sorori_shinzaemon(n_days=10)
```

```python
###             Cálculo y Visualización para 100 Días

# Cálculo para 100 días
N_DAYS = 100
list_n_grains_100, list_total_grains_100 = compute_sorori_shinzaemon(n_days=N_DAYS)

# Mostrar el resultado final (día 100)
print(f"\n--- Resultados para {N_DAYS} Días ---")
print(f"Granos recibidos el día 100: {list_n_grains_100[-1]:,}")
print(f"Granos TOTALES después de 100 días: {list_total_grains_100[-1]:,}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Gráfico 1: Granos recibidos CADA DÍA
plt.subplot(1, 2, 1) # 1 fila, 2 columnas, gráfico 1
plt.title("ある日に貰う米の数のリスト (リストの長さはn_daysに等しい)")
plt.xlabel("Día")
plt.ylabel("Cantidad de Granos (log)")
plt.plot(list_n_grains_100, color='blue', linewidth=2, alpha=0.7)
plt.yscale('log') # Eje Y en escala logarítmica
plt.grid(True, linestyle='--')

# Gráfico 2: Granos de Arroz TOTALES ACUMULADOS
plt.subplot(1, 2, 2) # 1 fila, 2 columnas, gráfico 2
plt.title("ある日までに累計で貰う米の数のリスト (リストの長さはn_daysに等しい)")
plt.xlabel("Día")
plt.ylabel("Cantidad de Granos Acumulados (log)")
plt.plot(list_total_grains_100, color='red', linewidth=2, alpha=0.7)
plt.yscale('log') # Eje Y en escala logarítmica
plt.grid(True, linestyle='--')

plt.tight_layout() # Ajusta automáticamente los subgráficos
plt.show()
```
